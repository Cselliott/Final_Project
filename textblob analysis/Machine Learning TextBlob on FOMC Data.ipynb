```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from textblob import TextBlob
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
#% matplotlib inline
```

```python
data = []
data_labels = []
with open("fomc/FOMC2007meeting.csv") as f:
    for i in f: 
        data.append(i) 
        #data_labels.append('pos')
        
# Converts the list object of the dataFrame into a string to be manipulated by other code.
newString = "".join(data)
meeting_text = newString

# Create blob_object
meeting_blob = TextBlob(meeting_text)
```
