```python
# Import modules
from bs4 import BeautifulSoup as bs
from splinter import Browser
import os
import pandas as pd
import time
```

```python
# Executable path is where chromedriver.exe exists
executable_path = {"executable_path":"c:\chromedriver\chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless = False)
```

```python
# Launch the web page
url = "https://www.federalreserve.gov/monetarypolicy/fomchistorical2007.htm"
browser.visit(url)
time.sleep(2)

# Use BeautifulSoup to write it into html
html = browser.html
soup = bs(html,"html.parser")

print(soup.prettify())
```

```python
import re
hrefs = soup.find_all("a", href=re.compile("meeting.pdf"))

import wget
import tabula
from tabula import read_pdf

for href in hrefs:
    link = "https://www.federalreserve.gov" + href.attrs['href']
    print(link)
    
    pdfname = link.split("FOMC")
    #print(pdfname[1])
    
    wget.download(link, "Resources/" + pdfname[1])
```
