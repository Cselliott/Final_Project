```python
import nltk
from nltk.tokenize import word_tokenize

event_text = "".join(df['event'][2])
```

```python
from nltk.corpus import stopwords

#stop_words = set(stopwords.words('english')) 

stopwords = nltk.corpus.stopwords.words('english')

newStopWords = ["There"] + ["\'s"] + list(stopwords)

#stopwords.union(newStopWords)
  
word_tokens = word_tokenize(event_text) 
  
filtered_sentence = [w for w in word_tokens if not w in newStopWords] 

print(event_text)
print()
print(word_tokens)
print()
print(filtered_sentence)
```

```python
df = pd.read_csv(r'../cleaned/FOMC20080130meeting_cleaned.csv') 

df = df.rename(columns={"meeting of the federal open market committee on january 2930 2008":"text"})

#df['text'][109]

df_recession = df[df['text'].str.contains("recession")] 
print(df_recession)
```

```python
stopwords = nltk.corpus.stopwords.words('english')

newStopWords = ["There"] + ["\'s"] + list(stopwords)

for index, row in df_recession.iterrows():
    #print(f"Index {index}")
    print(row['text'])
    print()
    
    word_tokens = word_tokenize(row['text'])

    filtered_sentence = []

    for w in word_tokens: 
        if w not in newStopWords: 
            filtered_sentence.append(w) 
            
    print(filtered_sentence) 
    print()
```
