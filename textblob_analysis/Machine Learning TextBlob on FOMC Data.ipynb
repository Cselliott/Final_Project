```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from textblob import TextBlob
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
data = []
data_labels = []
with open("fomc/bigdata.csv") as f:
    for i in f: 
        data.append(i) 
        
# Converts the list object of the dataFrame into a string to be manipulated by other code.
newString = "".join(data)
meeting_text = newString

# Create blob_object
meeting_blob = TextBlob(meeting_text)
```

```python
# Each word in the lexicon has scores for:
# 1) polarity: negative vs. positive    (-1.0 => +1.0)
# 2) subjectivity: objective vs. subjective (+0.0 => +1.0)
# 3) intensity: modifies next word?      (x0.5 => x2.0)


# The polarity score is a float within the range [-1.0, 1.0].
# Polarity in sentiment analysis refers to identifying sentiment orientation (positive, neutral, and negative) 
# in written or spoken language

# The subjectivity is a float within the range [0.0, 1.0] 
# where 0.0 is very objective and 1.0 is very subjective.

meeting_blob.sentiment
```

```python
meeting_blob.word_counts
```

```python
text = meeting_blob
stop_words = ["see","thing","going","well","now","will","might","say","much","lot","view","come","want","even","given","seem","actually","chairman","bernanke","point","right","think","bank","mr","one","market","look","things","dont","way","time","may","funds","rate","term","still","need","percent","us","issue","thats","really"]+ list(STOPWORDS)
wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    max_words=50,
    background_color = 'black',
    stopwords = stop_words).generate(str(text))
fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
